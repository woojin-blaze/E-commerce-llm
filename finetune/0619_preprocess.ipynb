```python
import re


def split_text_based_on_keyword(text):
    # Define the split keywords
    split_keywords = ["### Related Keywords:", "### Product Title:", "### Detailed Description:"]
    
    # Check for each split keyword
    for keyword in split_keywords:
        if keyword in text:
            parts = re.split(f"({keyword})", text, 1)
            before_keyword = parts[0]+ parts[1].strip()
            after_keyword = parts[2].strip() if len(parts) > 2 else ""
            
            # Clean the keyword
            cleaned_keyword = re.sub(r"^###\s*", "", keyword).strip()
            cleaned_keyword = re.sub(r":\s*$", "", cleaned_keyword).strip()
            
            return before_keyword, cleaned_keyword, after_keyword
    
    # If no keyword found, return the original text and empty string
    return text, "", ""


```

```python
base_test[['text', 'type', 'answers']] = base_test['text'].apply(lambda x: pd.Series(split_text_based_on_keyword(x)))


base_test
```

```python
import pandas as pd

# Define the search text
search_text = "2023 신제품 남성용 스마트워치 다기능 스포츠형 심박수 모니터링 블루투스 통화 방수 패션 비즈니스 시계"

# List of files to search in
files = [
    './data/base0619_test_split_1.csv',
    './data/base0619_test_split_2.csv',
    './data/base0619_test_split_3.csv',
    './data/base0619_test_split_4.csv'
]

# Loop through each file and search for the text
results = []
for file in files:
    df = pd.read_csv(file)
    matches = df[df['text'].str.contains(search_text, na=False)]
    
    if not matches.empty:
        for index, row in matches.iterrows():
            result = {
                'file': file,
                'row_index': index,
                'text': row['text']
            }
            results.append(result)
            print(f"Found in file: {file}, row: {index}")
            print(f"Text: {row['text']}\n")

# Check if any matches were found
if not results:
    print("No matches found.")
else:
    print(f"Total matches found: {len(results)}")

# Optionally, convert the results to a DataFrame and save to a file
results_df = pd.DataFrame(results)

```
