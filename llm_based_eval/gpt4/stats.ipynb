```python
import pandas as pd
import json

def load_json_data(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        return json.load(file)

def convert_to_dataframe(json_data):
    return pd.DataFrame(json_data)

def calculate_stats(dataframe, column_list):
    averages = dataframe[column_list].mean()
    variances = dataframe[column_list].var()
    
    return averages, variances
```

```python
d1=load_json_data("data/data3_base_0501_last.json")
```

```python
columns_to_analyze = [
    'score_pn_coher_detailed', 'score_pn_rel_detailed', 'score_pn_search_detailed',
    'score_cp_rel_detailed', 'score_cp_sen_comp_detailed', 'score_cp_harm_detailed',
    'score_rk_rel_detailed', 'score_rk_trend_detailed', 'score_rk_daily_detailed'
]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_stats(averages, variances, title):
    fig, ax = plt.subplots(2, 1, figsize=(10, 12))

    # Plot averages
    sns.barplot(x=averages.values, y=averages.index, ax=ax[0], palette='viridis')
    ax[0].set_title('Average Scores')
    ax[0].set_xlabel('Average Value')
    ax[0].set_ylabel('Score Categories')

    # Plot variances
    sns.barplot(x=variances.values, y=variances.index, ax=ax[1], palette='viridis')
    ax[1].set_title('Variance of Scores')
    ax[1].set_xlabel('Variance Value')
    ax[1].set_ylabel('Score Categories')

    plt.suptitle(title)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    
plot_stats(aver1, var1, 'Statistical Analysis of Product Scores')
```
